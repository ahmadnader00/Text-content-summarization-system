```python
from sumy.parsers.html import HtmlParser
from sumy.nlp.tokenizers import Tokenizer
```

```python
LANGUAGE = "arabic"
SENTENCES_COUNT = 3
url = "https://www.bbc.com/arabic/world-60982224"
parser = HtmlParser.from_url(url, Tokenizer(LANGUAGE))
for sentence in parser.document.sentences:
    print(sentence)
```

```python
from sumy.utils import get_stop_words

# Text to analyze
text = "This is a sample text that contains stop words "

# Get stop words for English language
stop_words = get_stop_words("english")

# Remove stop words from text
filtered_text = [word for word in text.split() if word.lower() not in stop_words]

# Print filtered text
print(filtered_text)
```

```python
from sumy.utils import get_stop_words
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("arabic")
stop_words = get_stop_words("arabic")

text = "هذا هو المشروع الفصلي "

words = text.split()
stemmed_words = []

for word in words:
    if word not in stop_words:
        stemmed_word = stemmer.stem(word)
        stemmed_words.append(stemmed_word)

print(stemmed_words)
```

```python
from sumy.nlp.tokenizers import Tokenizer

language = "arabic"
tokenizer = Tokenizer(language)

text = "أهلاً وسهلاً بكم في OpenAI. نحن نقدم حلولًا مبتكرة باستخدام الذكاء الاصطناعي."

# Using the NLP tokenizer from `sumy.nlp.tokenizers`
nlp_tokens = tokenizer._word_tokenizer.tokenize(text)

print(list(nlp_tokens))


```
